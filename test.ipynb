```python
import pandas as pd
import numpy as np
import xgboost as xgb

# Example input features
input_features = pd.DataFrame({
    'shop_id': [5],
    'item_id': [5037],
    'shop_category': [1],
    'shop_city': [2],
    'item_category_id': [3],
    'category_type': [4],
    'category_subtype': [5],
    'month': [6],
    'item_cnt_month_lag_1': [0],
    'item_cnt_month_lag_2': [0],
    'item_cnt_month_lag_3': [0],
    'item_cnt_month_lag_4': [0],
    'item_cnt_month_lag_5': [0],
    'item_cnt_month_lag_12': [0],
    'cnt_block_shop_lag_1': [0],
    'cnt_block_shop_lag_2': [0],
    'cnt_block_shop_lag_3': [0],
    'cnt_block_shop_lag_4': [0],
    'cnt_block_shop_lag_5': [0],
    'cnt_block_shop_lag_12': [0],
    'cnt_block_item_lag_1': [0],
    'cnt_block_item_lag_2': [0],
    'cnt_block_item_lag_3': [0],
    'cnt_block_item_lag_4': [0],
    'cnt_block_item_lag_5': [0],
    'cnt_block_item_lag_12': [0],
    'cnt_block_category_lag_1': [0],
    'cnt_block_category_lag_2': [0],
    'cnt_block_category_lag_3': [0],
    'cnt_block_category_lag_4': [0],
    'cnt_block_category_lag_5': [0],
    'cnt_block_category_lag_12': [0]
})

# Load the XGBoost model
model = xgb.Booster()
model.load_model("xgboost_model.json")

# Convert the input features into DMatrix for prediction
dmatrix = xgb.DMatrix(input_features)

# Make predictions
predictions = model.predict(dmatrix).clip(0, 20)

# Display the prediction
print(f"Predicted Item Sales: {predictions[0]:.2f}")
```
